```python
#min and max

a=int(input())
b=int(input())
c=int(input())
if a>b and a>c:
    print(a)
elif b>c and b>a:
    print(b)
else:
    print(c)
```

```python
a,b,c,d=map(int,input().split())
x=(a*b)-(c*d)
print("difference=",x)
```

```python
A,B,C,D=map(int,input().split())
print("difference=",(A*B)-(C*D))
```

```python
#Multiple
a,b=map(int,input().split())
if a % b == 0 or b % a == 0:
    print("multiples")
else:
    print("no multiples")
```

```python
#min and max
a,b,c=map(int,input().split())
print(min(a,b,c),max(a,b,c))
```

```python
#calculator

exp=input()
if '+' in exp:
    a,b=map(int,exp.split('+'))
    print(a+b)
elif '-' in exp:
    a,b=map(int,exp.split('-'))
    print(a-b)
elif '*' in exp:
    a,b=map(int,exp.split('*'))
    print(a*b)
elif '/' in exp:
    a,b=map(int,exp.split('/'))
    print(a//b)
    
```
