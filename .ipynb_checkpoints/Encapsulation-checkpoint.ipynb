### Encapsulation:
                    - Encapsulation means,wraping the methods and attributes in single unit that one is class.The main aim of Encapsulation is to restrict the direct access of attributes and methods.
                    - This type of retriction can implenment by using `Access Modifier`.
                    - Types of Access modifier:
                    1. public
                    2. protected
                    3. private

                    - In python we don't have any keywords to represent it public ,protected and private,but by using underscore(_) can define.

                    1. public:The public members(i.e attributes  + method) dosen't start with a under score.This public member can access any where in the program.
                    2. Protected:The protected members(i.e attributes  + method) dosen't start with a single under score.This members can access in inside the class and In subclasses also.
                    3.Private:The private members(i.e attributes  + method) dosen't start with a double under score.This members can access in only inside the class.




                    Tables:

                    Accessmodifier               Inside class       Sub class      Outside class
                    ..............................................................................            
                        Public                     yes                yes              yes
                        protected                  yes                yes              no
                        private                    yes                no               no

        1.Python does'nt support high security here we can access the protucted meb and private meb in outside the class by using following ways.
        to access the protucted mebmer 
        Syntax:
        (obj._followe by proticte name)
        2.To acces the private member by using name handling
            syntax:
                    (obj._classname__private member name)
        Advantages:
        the advantages is Encapsulation is the provided the security of the data as well as  functionalities.

```python
class Bank:
    def __init__(self,name,account,password,balance):
        self.name=name
        self.account=account
        self.__password = password
        self._balance = balance
    def total_users(self):                   # getter method
        return "Total users : 102"
    def get_balance(self):                     # getter method
        return f"current Balance is:{self._balance}"
    def deposite(self,deposite_amount):
        self._balance += deposite_amount
        return f" {deposite_amount} Deposited Successfully"
u1 = Bank("Rakesh", 2000,"Rakesh@123",10000)
print(u1.name)
print(u1.get_balance())  
print(u1.deposite(2000))
print(u1.get_balance())
print(u1._balance)           # protected  member access in outside the class
#print(u1._Bank__total_users())   # Nmae handling : syntax: obj._class_attributes)
```

### Inheritance

            Inheritance:In python,a class get properties from another class this process called as inheritance.

            1.A child class is inherited from parents class. 

                    Types of inheritance:
                                        1.Single inheritance
                                        2.Multi level inheritance
                                        3.Multiple inheritance
                                        4.Hierarchial inheritance
                                        5.Hybrid inheritance

                    1. Single inheritance:
                                        A child class get properties from single parent class.this is called as Single inheritance.

                                    ...............................................................................
                                          example:          parent           father            basic phone
                                                              ^                ^                   ^
                                                              |                |                   |
                                                            child            child             Smart phone
                                    ..............................................................................
                                          syntax:
                                                  class parent_class_name:
                                                          # attributes
                                                          # methods
                                                  class child class_name(parents_class_name):
                                                          # attributes
                                                          # methods

                    2. Multi level inheritance:A class (i.e child class) get properties from another class (i.e parents class) that class get properties from another class (i.e grand parent). 
                                 .......................................................

                                    example:
                                                    grand parents                     
                                                          |
                                                      parents
                                                          |
                                                        child


                                                        Animal
                                                           |
                                                          dog
                                                           | 
                                                          puppy
                                 ..........................................................

                    3.Multiple inheritance:A single class get properties from multiple parent classes. this is called as "multiple inheritance".

                                    ..........................................................
                                    example:  
                                             1.   (parent1)     (parent2)       (parent3) 
                                                        \                           /
                                                        ............................

                                                                   |
                                                                (child)


                                             2.     (mother)       (father)
                                                         \            /
                                                         .............
                                                               |
                                                            (child)
                                    ...........................................................


                    4.Hierarchial inheritance:Multiple child classes get properties from single parents class.

                                example:
                                            parent
                                            /  \  \
                                           c1  c2 c3


                    5.Hybrid inheritance:This is combination of two or more type of inheritances.

```python

#Single inheretance
# child class with constructor
# without child class implementation
# When child class don't have constructor in that case,parent class constructor behave like child class constructor
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def details(self):
        return f"my name {self.name} and i am a {self.age} years old gender {self.gender}"
        
class Student(Person):
    def __init__(self,stdid,name,age,branch,gender):
        super().__init__(name,age,gender)
        self.stdid=stdid
        self.branch=branch
    
s=Student(101,"Rakesh","Male",25,"CSE")
print(s.name)
print(s.branch)
print(s.gender)  # attributeserror:Student object has no attributes gender
print(s.details())  # it raise attributeerror because 
```

```python
# multi level inheritance
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def details(self):
        return f"My self {self.name}, and i am {self.age}"


class Student(Person):
    def __init__(self, name, age, branch):
        super().__init__(name, age)
        self.branch = branch


class Employee(Student):
    def __init__(self, name, age, branch, empid):
        super().__init__(name, age, branch)
        self.empid = empid
        
    def display(self):
        print(self.details())
        print(self.branch)
        print(self.empid)
e = Employee("Rakesh", 22, "CSE", 101)
print(e.name)
e.display()
```

```python
# Multi-Level Inheritance

class Animal:
    def __init__(self, animal_name):
        self.animal_name = animal_name

    def animal_details(self):
        print(self.animal_name)


class Dog(Animal):
    def __init__(self, animal_name, breed):
        super().__init__(animal_name)
        self.breed = breed

    def dog_details(self):
        print(self.breed)


class Puppy(Dog):
    def __init__(self, animal_name, breed, age):
        super().__init__(animal_name, breed)
        self.age = age

    def display(self):
        self.animal_details()
        self.dog_details()
        print(self.age)

p = Puppy("Tommy", "Labrador", 1)
p.display()
```

```python

# Multiple Inheritance:

class Father:
    def __init__(self, father_name):
        self.father_name = father_name

    def work(self):
        return f"Father is engineer"


class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

    def work(self):
        return f"Mother is teacher"


class son(Father, Mother):
    def __init__(self, age, name,father_name, mother_name):
        Father.__init__(self,father_name)
        Mother.__init__(self,mother_name)
        self.name=name
        self.age=age

    def parent_work(self):
        print(Father(self.father_name).work())
        print(Mother(self.mother_name).work())


s = son("srinu",20,"srinu","amma")
print(s.father_name)
print(s.mother_name)
print(s.work())
```

```python

# Multiple Inheritance:

class Father:
    def __init__(self, father_name):
        self.father_name = father_name

    def work1(self):
        return f"Father is engineer"


class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

    def work2(self):
        return f"Mother is teacher"


class son(Father, Mother):
    def __init__(self, age, name,father_name, mother_name):
        Father.__init__(self,father_name)
        Mother.__init__(self,mother_name)
        self.name=name
        self.age=age

    def parent_work(self):
        print(super().work1())


s = son("srinu",20,"srinu","amma")
print(s.father_name)
print(s.mother_name)
print(s.work2())
```

```python
# Hierarchical Inheritance

class Person:
    def __init__(self, name):
        self.name = name

    def details(self):
        print("Name:", self.name)


class Student(Person):
    def __init__(self, name, branch):
        super().__init__(name)
        self.branch = branch

    def student_info(self):
        print("Branch:", self.branch)


class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

    def employee_info(self):
        print("Salary:", self.salary)


s = Student("Rakesh", "CSE")
e = Employee("Naidu", 50000)

s.details()
s.student_info()

print()

e.details()
e.employee_info()
```
