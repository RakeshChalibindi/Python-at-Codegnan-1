#### Function:
                function is reuable block of code .in python function defination start with "def" keywords.
            syntax:
                    #function defination block
                    def fun_name(parameter:(optional):
                        #block code
                        return exp/value(optional)
                    #function calling block
                        fun_name(arguments:(optional))


Types of functions:
                    1.user defined function
                    2.builtin functions

        1.User defined function:
                                the function defination is developed by the user or programmer,those functions are called as a "user defined function"
                            ->a function used perform specific task
                            ->the function defination block execute whhen function call is occurs.
        parametrs:
                the function definations block,takes variable as input,those variable defined inside the paranthesis ,those variables called as "parameters".
                ->parameters are optional

        Arguments:
                    the function calling block task variables as inputs,those variable defined inside function calling paranthsis,those variable called as "arguments".


        Based on parameters and return values function are classified into 4 types.
                1.without paramets and without return
                2.without parameter and with return
                3.with parameter and without return
                4.with parameter and with return
                
                

```python
# Type of functions based on parameter and return values
# 4.with parameter and with return
#function defination block
def addition(x:int,y:int)->int:
    c = x+y
    return c 
# function calling block
a,b,=10,40
res=addition(a,b)
print(res)
```

```python
def subtraction(x:int,y:int):
    c=x-y
    return c
a,b=20,10
res=subtraction(a,b)
print(res)
```

```python
def multiplication(x:int,y:int):
    c=x*y
    return c
a,b=5,5
res=multiplication(a,b)
print(res)
```

```python
#3.with parameter without return
#by default function retuns none value,when function not returning any value
#function defination block

def addition(x:int,y:int)->None:
    c = x+y
     print(f"this is from function defination block:{c}")
# function calling block
a,b,=10,40
res=addition(a,b)
print(f"this is from function block:{res}")
```

```python

#without parameter and with return

def addition()->int:
    a=10
    b=33
    c=a+b
    return c
# function calling block
res=addition()
print(res)
```

```python
def subtraction()->int:
    a=10
    b=33
    c=a-b
    return c
# function calling block
res=subtraction()
print(res)
```

```python
#implement even or odd usinf function

def even_or_odd(x:int)->str:
    if x % 2==0:
        return "even"
    else:
        return "odd"
#function calling block
n=int(input())
print(even_or_odd(n))
```

```python
 #prime number using fraction

def prime_number(n:int)->str:
    if n<2:
        return " not prime"
    for i in range(2,n//2 +1):
        if n%i==0:
            return "not prime"
    return "prime"
#function calling 
n=int(input())
print(prime_number(n))
```
