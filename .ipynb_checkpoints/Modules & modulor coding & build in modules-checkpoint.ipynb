### lModules & Modular coding:
                            - A module is .py file and it consists of variable ,functions and classes.
                            - can use the one modules code in some another modules by importing the module.
        Advantages of modular codeing:
                                        1.code reusability
                                        2.easy to maintain
                                        3.easy to debug
                                        
            

### Build in modules:
             Types of modules:
                             1.user defined modules
                             2.build in modules
            1.user defined modules:
                                    The modules are defined by the user or programmer,those modules called as user defined modules.


            2.Build in modules:
                                the module are provided by the python software,t hose modules called as build in modules.
                            ex:
                               - math--->Advanced math operations.
                               - random--->To generate random numbers.
                               - os--->operating system-files,directory creations.
                               - sys--->system--related to interpreted .
                               - pandas--->data maniplations perpous.
                               - numpy--->Numerical calculations.
                               - matplotlib--->for data visulizations.
                               - scikit-learn--->for machine learning.

```python
import math
print(math.ceil(4.33))   #it return the upperbound integer value
print(math.floor(4.25))  #it return the lowerbound integer value
print(math.factorial(5)) # it  return the factorial of a number
print(math.gcd(10,6,4))  # It return the gratest  common devisor
print(math.lcm(10,6,4))  # It return the least common multiplication
print(math.pow(15,2))    # pow(a,b) :It return the float type of b raise to a.
print(math.sqrt(12))     # It return the squareroot values of a number
print(math.log10(5))     # to find the log10 values
```

```python
# import random function
import random
print(random.random())   # This function generate the random number between  0 to 1
print(random.randint(1,6500))   # this functiomn generates the random number between a range
print(random.randint(10000,63442))  # 4 digitotp generations
print(random.choice(['R',1,2,3,4,5,6,'S','E','F']))  #This function return the random values from sequence.
# print(random.choices(['R',1,2,3,4,5,6,'S','E','F'], F=25))  #This function return the random values from sequence.
arr = ['R',1,2,3,4,5,6,'S','E','F']
random.shuffle(arr)
print(arr)
```

```python
# who will reach the top(100) in less steps
import random
count=roles=0
print("Wellcome to the who will reach top in less steps Games")
while True:
    input("To trow a dice,plase click enter button")
    face_val=random.randint(1,6)
    if count + face_val <= 100:
        count += face_val
    roles+=1
    if count == 100:
        print(f"your dice values:{face_val}")
        print(f"You reached the top in {roles} changes")
        break
    print(f"your Dice value:{face_val}")
    print(f"your at {count} position in {roles} changes")
    print("*" * 50)
```
