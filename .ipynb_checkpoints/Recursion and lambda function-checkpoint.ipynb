### Recusrion and lambda function:
    Recursion:A Function calls itself either directly or indirectly.
    Types of recursions:
                        1.Direct recursion
                        2.indirect recursion
    1.Direct recursion:A function calls itself until it reaches to base conditions.
            main blocks:
                        1.basic condition
                        2.recursive call
            ex:def fun(x)
                  x=x-1
                  return fun(x)
                fun(5) 
        1.Base conditions:The base condition is used to stop the infinate recursive call.
        2.recursive conditions:The function call recursively until it reaches to base conditions.
    
    2.Indirected recursive:Function "a" calls function is and function "b" calls function 
                

```python
def fun(x):
    if x==0:
        return 0
    x=x-1
    fun(x)
    print(x)
fun(5)
```

```python
def fun(x):
    if x==0:
        return 0
    fun(x-1)
    print(x)
fun(3)
```

```python
#print 1 to n using recursive
def fun(x):
    if x==0:
        return 0
    print(x)
    fun(x-1)
fun(3)
```

```python
#print  factorial methods
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n * factorial( n - 1)
n=int(input())
factorial(5)
```

```python
# find the sum of digits in a number
#n=532,o/p:10
def sumofdigits(n):
    if n==0:
        return 0
    return n% 10 + sumofdigits(n//10)
print(sumofdigits(532))
     
```

```python
#length
def length_of_no(x):  #definations
    length=len(x)
    return length
x=input()          #function call
length_of_no(x)
    
```

#### Lambda functions (or) Anoymous Functions:
                                            The function doesn't have function name those functions called as anonymous functions
                                            1.The anonymous functions can implements with "lambda"keyword
                                                Syntax:
                                                    lambda arguments:Expression
#### Single line if-else:
                            Syntax: 
                                    true-exp if condition else false-exp

```python
#find the square of a number by using lambda function
#by using single line function
s=lambda x:x**2
print(s(9))
```

```python
#find the square of list of functions
arr=[1,2,3,4,5,6,7,8]
print(list(map(lambda x: x**2,arr)))
```

```python
arr=[1,2,3,4,5,6,7,8]
print(list(filter(lambda x: x%2==0,arr)))
```

```python
#Single line if-else:
#                   Syntax: 
#                          true-exp if condition else false-exp  
x=15
res="even"if x%2==0 else "odd"
print(res)
```

```python
#find the frequency of each character
string=input()
count={}
for char in string:
    if char in count:
        count[char] += 1
    else:
        count[char] = 1
print(count)
```

```python
#find the frequency of each character by using sort cut methods
string=input()
count={}
for char in string:
    count[char]=count.get(char,0)+1
print(count)
```
