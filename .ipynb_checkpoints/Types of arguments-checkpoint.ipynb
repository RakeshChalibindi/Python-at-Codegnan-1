### Tyeps of arguments:
                    1.positional arguments
                    2.keywords arguments
                    3.default arguments
                    4.variable length of positional arguments    #arbitary
                    5.variable length of keyword arguments

    
                    

```python
#Type of arguments:
#1.position arguments:here arguments are assigns to parameter as per the sequential 
#function defination block

def sub(x:int,y:int)->int:    # x,y=positional arguments
    c=x-y
    return c
#function calling block
a,b=11,33
print(sub(a,b))
print(sub(b,a))
```

```python
#Keyword arguments:arguments are explicitly passed with parameter names
#as par the key word parameters

def sub(x:int,y:int)->int:    # x,y=positional arguments
    c=x-y
    return c
#function calling block
a,b=11,33
print(sub(x=a,y=b))
print(sub(y=b,x=a))
```

```python
def add(x:int,y:int)->int:    # x,y=positional arguments
    c=x+y
    return c
#function calling block
a,b=11,33
print(add(x=a,y=b))
print(add(y=b,x=a))
# print(add(10,x=6)) # it will raise the error
# print(add(x=12,y))  # positional error
print(add(5,y=16))
```

```python
def add(x:int,y:int,z:int)->int:
    c1=x+y+z
    return c1
a,b,c=11,22,33
print(add(x=a,y=b,z=c))
```

```python
def add(w:int,x:int,y:int,z:int)->int:
    r=w+x+y+z
    return r
#a,b,c,d=map(int(input().split())
a,b,c,d=10,20,30,40
print(add(w=a,x=b,y=c,z=d))
```

```python
#3.default arguments:the default arguments are optional arguments,if you didn't 
#pass values for default arguments,in that case it assign default values otherwise,it assign new arguments value
def add(w:int,x:int,y:int=0,z:int=0)->int:
    res=w+x+y+z
    return res
a,b,c,d=map(int, input().split())
print(add(a,b))
print(add(a,b,c))
print(add(a,b,c,d))
            
```

```python
def mul(w:int,x:int,y:int=0,z:int=0)->int:
    res=w*x*y*z
    return res
a,b,c,d=map(int, input().split())
print(mul(a,b))
print(mul(a,b,c))
print(mul(a,b,c,d))
```

```python
def mul(w:int,x:int,y:int=1,z:int=1)->int:
    res=w*x*y*z
    return res
a,b,c,d=map(int, input().split())
print(mul(a,b))
print(mul(a,b,c))
print(mul(a,b,c,d))
```

```python
#variable length and positional arguments
def add(x,y=0,*args):
    print(type(args))
    print(args)
    s=0
    for i in args:   # i==values
        s +=i
    return i
print(add(10,20,1,2,3,4,5))
```

```python
#variable length keyword arguments:
#*->positional arguments
#**->keyword arguments

def add(x,y=0,*args,**rakesh):
    print(type(rakesh))
    print(rakesh)
    print(rakesh.values())
    print(rakesh.keys())
print(add(10,20,2,3,4,5,6,name="rakesh",age=20,course="pfs"))
```

```python
#In python every type function calling block shares the arguments refference to parametre 
#if the arguments are immutable data types it,behave likes pass by value and if arguments are mutable datatypes(i.e list,set,dict) then behave likes pass by reference
#pass by values
def fun(x):
    print(id(x))
    x += 2
    print("after add 2")
    print(id(x))
    print(x)
a=20
print(id(a))
fun(a)
print(a)
```

```python
def fun(x):
    print(id(x))
    x += 2
    print("after add 2")
    print(id(x))
    print(x)
a=20
print("before function is execution 'a' values:",a)
fun(a)
print("after function is execution 'a' values:",a)
```

```python
#pass by reference

def add_ele(lst,ele):
    lst.append(ele)
arr=[1,2,3,4]
print("before function execution arr values:",arr)
add_ele(arr,150)
print("after function execution arr values:",arr)
```
