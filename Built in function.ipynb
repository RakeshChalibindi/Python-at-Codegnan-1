```python
#find the difference between two strings
s1="CODE"
s2="GNAN"
n=len(s1)
s=""
for i in range(n):
    diff = abs(ord(s1[i]) -ord(s2[i]))
    if diff == 0:
        s += s1[i]
    else:
        s += chr(64+diff)
print(s)
```

```python
#find bigger and smaller characters from two string
#compare each character in both strings and if s1[i] > s2[i] add s1[i] to bigger and s2[i] to
#if both same leave it

s1,s2=map(str,input().upper().split())
n=len(s1)
big =small=""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i] > s2[i]:
            big += s1[i]
            small += s2[i]
        else:
            big += s2[i]
            small += s1[i]
print(big)
print(small)
```

```python
#Map(fun,iterable):It applies the functional try on each elements in iterable
#fulter(fun,iterable):It filters the itables,it retuens the values which is not None
#find the squares of the list of integers
def squere(x):
    return x**2
lst = [1,2,3,4,5]
print(list(map(squere,lst)))
```

```python
#filter the even numbers from list of integers
def even(x):
    if x%2==0:
        return x
arr=[1,2,3,4,5,6,7,8]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
def even(x):
    if x%2==1:
        return x
arr=[1,2,3,4,5,6,7,8]
print(list(filter(even,arr)))
```

```python
def even(x):
    if x%2==0:
        return 1
arr=[1,2,3,4,5,6,7,8]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
#adding of two numbers
def squere(t):
    x,y = t
    return x+y
lst = [(1,2),(3,4)]
print(list(map(squere,lst)))
```

```python
#find the target elements position in list
arr=[10,20,30,40,50]
target=50
for ind,val in enumerate(arr):
    if target == val:
        print(ind)
        break
else:
    print(-1)
```
