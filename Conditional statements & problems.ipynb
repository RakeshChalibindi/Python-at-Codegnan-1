### Loop statements:
                    1.continue:after executing of 'continue' keyword execution immediately it goes to next iteration and skips the current iteration after continuous keyword statements.

```python
#loop control statements
#continue
for i in range(1,10):
    if i%2==0:
        continue
    print(i)
```

#### 2.break:
            executing of 'break' keyword execution immediately goes to out of loop,-that means loop will terminate -the loops else block execute when break keyword not executed.
