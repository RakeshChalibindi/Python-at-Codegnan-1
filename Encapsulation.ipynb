### Encapsulation:
                    - Encapsulation means,wraping the methods and attributes in single unit that one is class.The main aim of Encapsulation is to restrict the direct access of attributes and methods.
                    - This type of retriction can implenment by using `Access Modifier`.
                    - Types of Access modifier:
                    1. public
                    2. protected
                    3. private

                    - In python we don't have any keywords to represent it public ,protected and private,but by using underscore(_) can define.

                    1. public:The public members(i.e attributes  + method) dosen't start with a under score.This public member can access any where in the program.
                    2. Protected:The protected members(i.e attributes  + method) dosen't start with a single under score.This members can access in inside the class and In subclasses also.
                    3.Private:The private members(i.e attributes  + method) dosen't start with a double under score.This members can access in only inside the class.




                    Tables:

                    Accessmodifier               Inside class       Sub class      Outside class
                    ..............................................................................            
                        Public                     yes                yes              yes
                        protected                  yes                yes              no
                        proivate                   yes                no               no
