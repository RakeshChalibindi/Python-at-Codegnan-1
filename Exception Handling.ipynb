### Exception Handling

            Exception Handling:
                                Exception is nothing but it is an error.Whenever an exception is occurs it stops the program exections and it return type of exception and followed by exception messages.

                    Types of errors:
                                    1.Syntax error or compilertime error   
                                    2.Runtime error 
                                    3.Logical error

                    1.Syntax Error:
                                    The syntax errors raise during the compilation time of the program.Those error call it as Syntax error.

                                    Example:
                                            IntendationsError,SyntaxError.
                    2.Runtime Error:
                                    The Runtime errors raises durings Runtime of a program,Those error called as Runtime Error.
                                -   we can't handles syntax Error ,but we can handle the Runtime errors by using Exception handling blocks(i.e.try-except)

                                    Example:
                                            KeyError,ValueError,ZeroDivisionError,IndexError,TypeError,etc....


    Exception handling blocks:
                     Syntax:
                            try:
                                # block code that might contains exception.
                            Except:
                                # Exception handling block
                            else:[Optional]
                                # This block executes there  is not exception raised in try block
                            finally:[optional]
                                # This block always executes error exception raised or not

```python
# try block with exceptions

try:
    list=[1,2,3]
    print(list[10])
    print("hii")
except:
   print("something wrong in tryblock")
else:
    print("No exception raised in try block")
finally:
    print("program is done")
```

```python
#It  execute the specific exception block when specified exception raised otherwise,it executes the common except block
try:
    a=10
    b=a.append(1)
    c=a/b
    print(c)
except ZeroDivisionError as e:
    print("Zero division is not possiable: {e}")
except ValueError as e:
    print(f"Value Error raised in Try block: {e}")
except Exception as e:  # Common exception block with exception class
    print(f"something wrong in program:{e}")
```
