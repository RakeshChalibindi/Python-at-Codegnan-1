### Generators:
                Generator is a special types of function.
    It allows you to generate one values at time.
                1.The generator function returns a iterable object
        Syntax:
                def generate_name(parameters):
                    #block code
                    yield val/exp
                    yeild val2
                    yeild val 3
                res=generator_name(arguments):
                print(next(res))

    1.The generator function start exectues whenever if you call generator
    2.The generator function start exectue whenever if you call generator object with next() function.
    Then generator function defination start exectutes until it reduces to yeild statements and exectution is passed.

    3.If you call generator object again with next(),if start executing previosly where leftoff.
    4.Generator function is a "lazy" function exections
    5.more memory efficient
   

```python
def generator(n):
    for i in range(1,n+1):
        yield i
n=int(input())
res=generator(n)   #It doesn't start exectue generator 
print(next(res))   #Here generator function exectution start ,It exectues until it reduce to yield statements
print(next(res))
```
