```python
list=[1,2,3,4]
for i in list:
    if i%2==0:
        print("even")     #print(f"{i} is even")
    else:
        print("odd")      #print(f"{i} is odd")
```

```python
lst=list(map(int,input().split()))
res=[]
for i in lst:
    if i%2==0 and i%5==0:
        res.append("fizzbuzz")
    elif i%2==0:
        res.append("fizz")
    elif i%5==0:
        res.append("buzz")
    else:
        res.append("i")
print(res)
```

### for loop:

                        range(start,end,step)  #start value should be (0)

```python
for i in range(1,10):
    print(i)
```

```python
n=int(input())
for i in range(1,n+1):
    print(i,end=" ")     
```

### while: 
        The loop body executes until conditions evaluates false:
        1.In while loop munnal increments/decrements
        2.If you don't know the total no of iterations ,we will use while loop.

                Syntax:
                            #initilizatin 
                            while condition:
                                #block code
                                #increment/decrement
                            else:(optional)
                                #blockcode

```python
n=int(input())
for i in range(1,n+1):
    print("*",end=" ")
```

```python
#print 1 to n number by using while loop
n=int(input())
i=0
while i<=n:
    print(n)
    n-=1    #decerements
```

```python
n=int(input())
i=0
while i<=n:
    print(i)
    i+=1     #increments
```
