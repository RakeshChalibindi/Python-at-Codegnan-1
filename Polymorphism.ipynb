#### Polymorphism
            Polymorphism:In python polymorphism  `poly` means `many` and morphism means `forms`.
        1. In polymorphism single function it shows different behaviours based on input context.
        2. In python polymorphism can achive through the following ways.

                1.Through operators
                2.polymorphism  through functions
                3.polymorphism through methods
                            |
                            1.runtime polymorphism      ----------- Method overriding
                            2.compiler time polymorphism----------- Method overloading

        1.polymorphism through operations:
                                            Here single operator it shows different actions based on input context.
                                                For example  '+' operator it perform and addition operations when inputs are integers as well as it forms string concatenation when input are string.

                                                        ex:
                                                            3+5=8  addition
                                                            "code"+ "gnan" = string concatenations..

                                                            3*5=multiplication operation

        2.Polymorphism through functions:
                                        Here a function shows different actions based on input context.

                                            ex:
                                                len("hello")=string length
                                                len(([1,2,3,4])=list length

        3.Polymorphism through methodS:
                1.Method overriding:(Throught the inheritance)
                                    A method defined in child class,that is exist in parent class with the samename   .
                                    Whenever if you call that method throught child class object,the child class methods overrides the parent class method.This is know as method overriding.

                                    Syntax:
                                            class a:
                                                def show()
                                            class b:
                                                def show()
                                            b=b()
                                            b.show()
                                            
                            
            

```python
# Method overriding
# Even if you define differnet no parements in method,always child method overide parent class method.
class Person:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender
    def details(self):
        return f"My name is {self.name} and i am a {self.age} year old"
class Student(Person):
    def __init__(self,stdid,name,age,gender,branch):
        super().__init__(name,age,gender)
        self.stdid=stdid
        self.branch=branch
    def details(self):
        return f"my name {self.name} and i am belong to {self.branch} branch"
S=Student(20,"Rakesh",33,"male","cse")
print(S.details())
    
```

```python
#Method Overloading:

# A class contains multiple methods with same name with different parameter list,in that case last defined method overrides the previous defined methods.This called as Method overloading
# Directly python doesn't support method overloading but we can achived same functionality through default arguments,variable length positional,or variable length keywords arguments
```

```python
# Method overloading

class Math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d
m=Math()
print(m.add(1,2,3,4))
# print(m.add(1,2,3))
# print(m.add(1,2))


```

```python
# Method overloading
# Directly python doesn't support method overloading but,we can achive same functional through the default argument or Arbitary arguments.

class Math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c=0,d=0):
        return a+b+c+d
m=Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))


```

```python
#  Magic methods or special methods

class Point:
    def __init__(self,x):
        self.x=x
p1=Point(5)
print(p1)
```

```python
# String 
#__str__() Magic method
class Point:
    def __init__(self,x):
        self.x=x
    def __str__(self):
        return f"This is point class and point value is {self.x}"
    def __add__(self,other):
        return self.x + other.x
p1=Point(5)
print(p1)
p2=Point(3)
print(p1.x + p2.x)
print(p1 +p2)
```

```python
# String 
#__str__() Magic method
class Point:
    def __init__(self,x):
        self.x=x
    def __str__(self):
        return f"This is point class and point value is {self.x}"
    def __add__(self,other):
        return Point(self.x + other.x)
p1=Point(5)
print(p1)
p2=Point(3)
# print(p1.x + p2.x)
print(p1 +p2)
p3=p1+p2
print(type(p3))
print(p3)
```

```python
class Number:
    def __init__(self,x):
        self.x=x
    def __add__(self,y):
        return self.x + y.x
    def __sub__(self,y):
        return self.x - y.x

    def __mul__(self,y):
        return self.x * y.x
n1=Number(20)
n2=Number(70)
print(n1+n2)
print(n1-n2)
print(n1*n2)
        
```
