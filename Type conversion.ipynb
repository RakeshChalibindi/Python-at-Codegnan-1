#### Type conversion:
                    Converting one data type format into another datatype formate.
                    This is called as type conversion

            They are 2 types:
                        1.implicit type conversion
                        2.Explicit type conversion(type casting)

            1.Implicit type conversion:
                                        The python interpreter automatically converts the datatype because ,it prevents the data loss.
                                        1.Implicit type conversion can happends in arithematic operations

                                                Example:
                                                        3+5.8=8.8
            2.Explicit type conversion(type casting):
                                                    Here type conversion happends by using python built in datatype function
                                         example:   a=2.3
                                                    b=int(a)
                                                    print(b)

            Python built in datatype function:
                        1.interger-int()
                        2.float val-complex()
                        3.boolean -bool()
                        4.string-str()
                        5.list-list()
                        6.tuple-tuple()
                        7.set-set()
                        8.dictionary-dict()

            Type conversion:
                               1. integer-float,string,complex,boolean
                               2. float-int,bool,str
                               3. Bool-int,

```python
#integer to float,boolean,string

num=10
print(float(num))
print(bool(num))
print(str(num))  #Same value should be print 
print(complex(num))
```

```python
number=12,33,444
print(list(number))
print(set(number))
```

```python
#float to int,bool,string

val=10.245
num=int(val)
print(type(num),num)
print(bool(val))
str_val =str(val)
print(type(str_val),str_val)
print(str_val[:2])
print(str_val[2],str_val[-2])
```

```python
#String to int,float,bool,list,tuple,set
#if string values is number

s="123456789"
print(type(s))
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```
