```python
# define person class
class Student:
    school=" xyz school"          #class attributes  outside the constructor
    def __init__(self,name,age):  #(__ magic method or specific methods)# Initialization constructor   #self --object represent/instance.
        self.name=name             #instance attributes 
        self.age=age              #instance attributes   #statemnts means attributes
        #print("This Student class __init__method")
    def details(self):
        return f"My name is {self.name} and i am a {self.age} year old"
        
s1=Student("rakesh",22)
s2=Student("srinu",45)
print(s1.name)
print(s2.age)
print(s1.details())
print(s1)
print(s1.school)
```

 ### Methods:
     Methods:The function is implemented inside the class those function called as Methods
             Types of methods:
                             1.Instance method
                             2.class method
                             3.ststic method
                             
        -    1.Instance method:The method defination takes first parameter as 'self' those methods called as 'Instance Method'. <br> This methods works on instance attributes.
                            
                                Synatx: 
                                        class class_name:
                                            def method_name(self):
                                                #block code
            2.Class method:
                        The method defination start with class with 'classmethod' decorator and it takes first parameter as 'cls' those methods called as 'class methods'.this methods works on class attributes.
                        
                                Syntax: `<br>`
                                        class class_name:  <br>--break
                                            @classmethod
                                            def method_name(cls):            <br>
                                                #block code



            3.Static methods:
                            The method defination start with `Staticmethod` decorator and it does't take first parameter as `self or cls` those methods called as `statics method`.this method works on except instance attributes and class attributes.
                            
                                    
                                Syntax:
                                        class class_name:  <br>--break
                                            @staticsmethod
                                            def method_name(parameters):            
                                                #block code
            4.the intialization constructor is used for to intialiaze the instance attribute data and this constructor takes first parameter as self and this constructor executes whenever an object is created for class
