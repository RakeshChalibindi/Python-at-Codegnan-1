#### Scope of variables:
                        A scope represents region of a program and where variable or name can access.
        Types of scopes:
                        1.local scope
                        2.enclosing scope
                        3.global scope
                        4.built in scope
       1.local scope:define a varible inside the function.
               ex:
                   def fun()
                       x=10   #locla variable
                       print(x)
                    fun()


        2.enclosing scope:a variable is defined in outer function and accessing in inner function (nested function).
                    ex:
                        def outer():
                            y=10    #enclosing variable/scope
                            def inner():
                            z=22    #local variable/scope
                            print(z,y)
                        inner()
                    outer()

                output:22,10

                
        3.global variable:outside all function
                    ex:
                        x=10           #global scope variable
                        def outer():
                            y=20       #enclosing scope avariable
                            def inner():
                                z=33   #local scope variable
                                print(x,y,z)
                            inner()
                        outer()


                    OUTPUT:33,20,10
        4.build in scope:the python provide the build in functions like print(),min(),max(),len(),find(),etc..


        note:When ever if you try to access any variable or name it start check in 'LEGB' rule
                LEGB MEAN:
                        L-local scope
                        E-enclosing scope
                        G-global scope
                        B-build scope
                        

```python
x=10  #global variable
def fun():
    print(x)
fun()
```

```python
x=20   #global scope
def outer():
    y=22   #enclosing scope
    def inner():
        z=33  #local scope
        print(z,y,x)   #print()--buildin in function
    inner()
outer()
```

```python
#using global scope variable in local scope
x=20   #global scope
def outer():
    y=x+1
    print(y)
outer()
print(x)
```

```python
#updating the global scope variable values in local scope
#can't updated global scope variable in local scope directly

x=20   #global scope
def outer():
    x=x+1
    print(y)
outer()
print(x)
```

```python
#updating the global scope variable values in local scope by using 'global' keyword

x=20   #global scope
def outer():
    global x
    x=x+1
    print(x)
outer()
print(x)
```

```python
#updating the enclosing scope variable value in local scope

def outer():
    x=20
    def inner():
        nonlocal x
        z=30
        x=x+z
        print(x,z)
    inner()
outer()
    
```

```python
#updating the enclosing scope variable value in local scope by using nonlocal

def outer():
    x=20   #enclosing
    def inner():
        nonlocal x
        z=30
        x=x+z  #local scope variable
        print(x,z) #print()-buildin scope function
    inner()
    print(x)
outer()
    
```
